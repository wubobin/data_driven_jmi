# 直接用108个数据拟合DFT带隙

```python
#注意参数不同
nfeat_bond = 25
r_cutoff = 6
gaussian_centers = np.linspace(0, r_cutoff+1, nfeat_bond)
gaussian_width = 0.5
graph_converter = CrystalGraph(cutoff=r_cutoff)
model = MEGNetModel(graph_converter=graph_converter, centers=gaussian_centers, width=gaussian_width)
```

```python
model.train(new_structures, new_dftbandgaps, epochs=200)
```

```python
# 使用模型预测能带差
predictions_train = []
for structure in new_structures:
    prediction = model.predict_structure(structure)
    predictions_train.append(prediction)
```

```python
# 计算均方根误差、平均绝对误差和R²,# 计算MAE
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score  

rmse_train = np.sqrt(mean_squared_error(new_dftbandgaps, predictions_train))
mae_train = mean_absolute_error(new_dftbandgaps,predictions_train)
r2_train = r2_score(new_dftbandgaps, predictions_train)
```

```python
rmse_train,mae_train,r2_train 
```

# 提取拟合对象

```python
import os
import glob

# 指定的路径
path = '/root/home/jupyter/dft_chgnet_m3gnet/CONTCAR_opt_dft_FAPbI2BrxI1-x/'

# 使用glob模块找到所有.cif后缀的文件
cif_files = glob.glob(os.path.join(path, '*.cif'))

# 打印结果
for file in cif_files:
    print(file)
```

```python
len(cif_files)
```

```python
pred_structures = []
for i in cif_files:
    structure = Structure.from_file(i)
    pred_structures.append(structure)
len(pred_structures)
```

```python
len(pred_structures)
```

```python
predictions_dftbandgaps = []
for i in cif_files:
    structure = Structure.from_file(i)
    prediction = model.predict_structure(structure)
    predictions_dftbandgaps.append(prediction)
```

```python
predictions_dftbandgaps_ = [pred[0].item() for pred in predictions_dftbandgaps]
predictions_dftbandgaps_
```

```python
# 遍历测试集的结构和预测gaps值，以及提取的DFT带隙值和真实gap值 
results = []
for structure, pre in zip(pred_structures, predictions_dftbandgaps_):  
    # 创建一个字典来存储当前结构的数据  
    item = {}  
    item['structure'] = structure.as_dict()  # 将Structure对象转换为字典  
    
    item['prediction_dftbandgaps'] = pre  # 直接使用预测gaps值，因为它已经是一个浮点数了  

    results.append(item)  

print(len(results))
```
