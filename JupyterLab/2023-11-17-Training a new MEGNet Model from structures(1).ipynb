```python
import numpy as np
from pymatgen.core import Structure
from megnet.models import MEGNetModel
from megnet.data.crystal import CrystalGraph
import json
```

```python
with open("/root/public/bandgap_json/DFT_bandgaptrain1.json", "r") as f:
    data_train = json.load(f)
```

```python
structures = []
gaps = []

for item in data_train:
    structure = Structure.from_dict(item['structure'])
    gap = item['real_gap'] - item['bandgap']
    structures.append(structure)
    gaps.append(gap)
```

```python
nfeat_bond = 10
r_cutoff = 8
gaussian_centers = np.linspace(0, r_cutoff+1, nfeat_bond)
gaussian_width = 0.5
graph_converter = CrystalGraph(cutoff=r_cutoff)
model = MEGNetModel(graph_converter=graph_converter, centers=gaussian_centers, width=gaussian_width)
```

```python
# Model training
# Here, `structures` is a list of pymatgen Structure objects.
# `targets` is a corresponding list of properties.
model.train(structures, gaps, epochs=150)
```

```python
# Predict the property of a new structure
model.predict_structure(structures[0])
```

```python
# 使用模型预测能带差
predictions = []
for structure in structures:
    prediction = model.predict_structure(structure)
    predictions.append(prediction)
```

```python
# 计算均方根误差、平均绝对误差和R²,# 计算MAE
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score  

rmse = np.sqrt(mean_squared_error(gaps, predictions))
mae = mean_absolute_error(gaps, predictions)
r2 = r2_score(gaps, predictions)
```

```python
print(r2)
print(mae)
print(rmse)
```

```python
# 导入模块    
import matplotlib.pyplot as plt    
import numpy as np    
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score  
  
# 更改默认字体为DejaVu Serif  
plt.rcParams['font.family'] = 'DejaVu Serif'  
  
# 绘制图像  
fig = plt.figure(figsize=(6,6),dpi = 600)  
plt.scatter(gaps, predictions, c='blue')  
plt.plot(np.linspace(-0.6, 0.65, 100), np.linspace(-0.6, 0.65, 100), linewidth=2, linestyle="--", color="black")  
plt.xlim(-0.6, 0.65)  
plt.ylim(-0.6, 0.65)  
plt.xticks(fontsize=18)  
plt.yticks(fontsize=18)  
plt.xlabel("Real_gap(eV)", fontsize=20)  
plt.ylabel("Prediction_gap (eV)", fontsize=20)  
  
# 在图上添加均方根误差、平均绝对误差和R²  
plt.text(-0.5, 0.5, 'R²: {:.2f}'.format(r2), fontsize=18)  
plt.text(-0.5, 0.4, 'RMSE: {:.4f}'.format(rmse), fontsize=18)  
#plt.text(-0.5, 0.3, 'MAE: {:.4f}'.format(mae), fontsize=16)  
  
# 在图上添加算法名称  
#plt.text(-0.5, 0.6, 'MEGNet', fontsize=16)  
plt.savefig('/root/public/Prediction_train1.jpg',bbox_inches='tight')  
plt.show()  
plt.close()
```
